```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
excel_file = pd.ExcelFile("VFTC-data-sheet.xlsx")
pagelist = []
for index in range(len(excel_file.sheet_names)):
  individual_page = pd.read_excel("VFTC-data-sheet.xlsx" ,sheet_name=index ,skiprows=2)
  pagelist.append(individual_page[['Name','Man. Date','Rate (%)']].dropna())

dataset = pd.concat(pagelist)
```

```python
X = dataset['Man. Date'].values.reshape(-1, 1) 
Y = dataset['Rate (%)'].values.reshape(-1, 1) 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
prediction = model.predict(X_test)
```

```python
plt.scatter(X_test, Y_test, color="black")
plt.plot(X_test, prediction, color="blue")
plt.xlabel("Age")
plt.ylabel("Opacity")
```

```python
input_age = int(input("Enter Age"))
```

```python
input_predection = model.predict([[input_age]])
plt.scatter(X_test, Y_test, color="black")
plt.plot(X_test, prediction, color="blue")
plt.xlabel("Age")
plt.ylabel("Opacity")
plt.plot(input_age,input_predection,'ro')
```
